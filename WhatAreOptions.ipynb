# Что такое опционы

### Что такое опцион

Опцион - это контракт, который дает право (но не обязательство) купить/продать базовый актив в заранее оговоренный момент времени по заранее оговорённой цене.

**Базовый актив** - это рассматриваемый актив, например акция либо криптовалюта.

Здесь важно, что у покупателя есть право, но не обязательство. То есть он в момент экспирации (истечения) опциона может и не исполнять этот опцион. Например, если этот опцион будет для него невыгоден.

В данной статьей будем рассматривать только **европейские опционы**, то есть те, которые можно исполнить только в момент истечения самого опциона.


### Опционы call и put
Есть два вида опционов - опцион на покупку **(call)** и опцион на продажу **(put)**.

Рассмотрим пример

Допустим человек купил опцион (call) на покупку эфира в размере 1 эфир по цене $\$1600$ (то есть со страйком 1600) через 30 дней.
Рассмотрим 2 случая:

- 1-ый: цена эфира через 30 дней опустилась ниже $\$1600$. В этом случае человеку не выгодно исполнять опцион (то есть покупать по $\$1600$), так как он может купить на рынке дешевле. Такой опцион называется **Out-of-the-money** ("не в деньгах")
- 2-ой: цена эфира через 30 дней стала значительно выше $\$1600$. Тогда ему становится выгодно исполнить опцион, купив 1 эфир по $\$1600$, так как он мгновенно может продать его на рынке дороже. 



#### В чём выгода для продавца опциона?

Продавец опциона берёт на себя обязательство исполнить опцион (то есть купить/продать себе в минус), если покупатель захочет исполнить опцион. Таким образом, продавец берёт на себя риск, и хочет получить премию. Премия за взятый риск - это и есть цена опциона.




### Как посчитать выплату по опциону

Выплата - то есть сколько мы заработаем/потеряем, если купим данный опцион.

Выплата по европейскому **call**-опциону на экспирации определяется формулой:

$$Payoff = max(0, S_T - K) - C_0$$

$S_T$ - цена базового актива на момент экспирации опциона <br/>
$K$ - страйк опциона <br/>
$C_0$ - цена call-опциона на момент покупки ($t=0$)

#### Выплата в зависимости от цены на момент исполнения опциона будет выглядеть так:



```python
K = 1500 # strike price
C0 = 25

prices = [price for price in range(500, 2500, 25)]
calls = [max(0, price - K) - C0 for price in prices]

plt.plot(prices, calls, 'b')
plt.title('Call option')
plt.xlabel('underlying price')
plt.ylabel('option payoff')
plt.show()
```

#### Для put-опциона:

Схожим образом считается выплата по европейскому **put**-опциону:

$$Payoff = max(0, K - S_T) - P_0$$

$S_T$ - цена базового актива на момент экспирации опциона <br/>
$K$ - страйк опциона <br/>
$P_0$ - цена put-опциона на момент покупки ($t=0$)



```python
K = 1500 # strike price
P0 = 45

prices = [price for price in range(500, 2500, 25)]
calls = [max(0, K - price) - P0 for price in prices]

plt.plot(prices, calls, 'b')
plt.title('Put option')
plt.xlabel('underlying price')
plt.ylabel('option payoff')
plt.show()
```

### Как посчитать стоимость опциона?

Здесь начинается самое интересное. Стоимость опциона - это премия за риск, который готов на себя взять продавец опциона. Причем цена теоретически может пойти совершенно любым образом. И в идеале хотелось бы иметь модель, которая бы говорила сколько конкретный опцион должен стоить (конечно, на практике идеально такое невозможно).

#### Почему это сложно?

Никто не знает как пойдёт цена в будущем. В зависимости от движения цены, будут получаться разные выплаты по опциону. Соответственно не понятно, как вообще искать цену опциона, которая будет устраивать продавца и покупателя.



#### Модель Блэка-Шоулза

Модель Блека-Шоулза позволяет аналитически вычислить честную стоимость опциона.
